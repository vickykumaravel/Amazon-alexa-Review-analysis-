# What is the overall sentiment of the reviews (positive, neutral, negative)? 

```python
def categorize_sentiment(rating):
    if rating >= 4:
        return 'Positive'
    elif rating == 3:
        return 'Neutral'
    else:
        return 'Negative'

# Apply sentiment categorization
data['sentiment'] = data['rating'].apply(categorize_sentiment)

# Calculate sentiment distribution
sentiment_counts = data['sentiment'].value_counts(normalize=True) * 100

# Plot sentiment distribution
plt.figure(figsize=(8, 6))
sentiment_counts.plot(kind='bar', color=['green', 'orange', 'red'], edgecolor='black')
plt.title('Overall Sentiment of Reviews')
plt.xlabel('Sentiment')
plt.ylabel('Percentage of Reviews (%)')
plt.xticks(rotation=0)
plt.show()

# Print sentiment distribution
print(sentiment_counts)
```

```python
# From the overall sentiment of the reviews we came to know that the positive reviews are 87.04%, negative reviews are 8.12% and 
neutral reviews are 4.82%
```

# What are the most common keywords in positive reviews?

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Load your data into a DataFrame (replace 'your_file.csv' with your actual file)
data = pd.read_csv("amazon_alexa.tsv",delimiter='\t')

# Assuming your DataFrame contains 'verified_reviews' and 'rating' columns
positive_reviews = data[data['rating'] >= 4]['verified_reviews']

# Use CountVectorizer to count word frequencies
vectorizer = CountVectorizer(stop_words='english')
word_count = vectorizer.fit_transform(positive_reviews)

# Sum up the counts of each word
word_sum = word_count.sum(axis=0)

# Get the words and their corresponding counts
words_freq = [(word, word_sum[0, idx]) for word, idx in vectorizer.vocabulary_.items()]

# Sort the words by frequency
words_freq = sorted(words_freq, key=lambda x: x[1], reverse=True)

# Get the top 10 most common words
most_common_words = words_freq[:10]

# Print the top 10 most common words
print(most_common_words)
```

```python
# These are the  top 10 most common keywords in positive reviews.
```

# What are the most common keywords in negative reviews?

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Load your data into a DataFrame (replace 'your_file.csv' with your actual file)
data = pd.read_csv("amazon_alexa.tsv", delimiter='\t')

# Drop rows with NaN values in the 'verified_reviews' column
data = data.dropna(subset=['verified_reviews'])

# Filter negative reviews
negative_reviews = data[data['rating'] <= 2]['verified_reviews']

# Use CountVectorizer to count word frequencies
vectorizer = CountVectorizer(stop_words='english')
word_count = vectorizer.fit_transform(negative_reviews)

# Sum up the counts of each word
word_sum = word_count.sum(axis=0)

# Get the words and their corresponding counts
words_freq = [(word, word_sum[0, idx]) for word, idx in vectorizer.vocabulary_.items()]

# Sort the words by frequency
words_freq=sorted(words_freq,key=lambda x: x[1], reverse=True)

# Get the top 10 most common words
most_common_words = words_freq[:10]
# Print the top 10 most common words
print(most_common_words)

```
